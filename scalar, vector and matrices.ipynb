 # Tensor

### Creating a tensor

```python
m1 = np.array([[5,12,6],
               [-3,0,14]])
m1
```

```python
m2 = np.array([[9,8,7],
               [1,3,-5]])
m2
```

```python
t = np.array([m1, m2])
t
```

```python
t.shape
```

### Adding Matrices

```python
m1 + m2
```

### Subtracting Matrices

```python
m1 - m2
```

### Transposing matrices

```python
a1 = np.array([[5,12,6],
               [-3,0,14]])
a1
```

```python
a1.T
```

### Dot product

```python
v1 = np.array([2,8,-4])
v2 = np.array([1,-7,3])
```

```python
np.dot(v1, v2)
```

```python
A = np.array([[5,12,6],
               [-3,0,14]])
A.shape
```

```python
B = np.array([[2,-1],[8,0],[3,0]])
B.shape
```

```python
A_x_B = np.dot(A,B)
A_x_B
```
