## Extract the data from CSV

```python
import numpy as np
from sklearn import preprocessing

raw_csv_data = np.loadtxt('Audiobooks_data.csv', delimiter=',')

unscaled_inputs_all = raw_csv_data[:,1:-1] # Takes all the columns except the first and last ones(id and targets)
targets_all = raw_csv_data[:,-1] # Assigning the last column of the csv into targets_all
```

## Balance the dataset

```python
num_one_targets = int(np.sum(targets_all)) # Counting the number of ones 

zero_targets_counter = 0
indices_to_remove = [] # Records the indices to be removed

for i in range(targets_all.shape[0]): 
    if targets_all[i] == 0:
        zero_targets_counter += 1
        if zero_targets_counter > num_one_targets: 
            indices_to_remove.append(i) # After the iteration, it will provide all the indices that won't be needed
            
unscaled_inputs_equal_priors = np.delete(unscaled_inputs_all, indices_to_remove, axis = 0) 
targtes_equal_priors = np.delete(targets_all, indices_to_remove, axis = 0)
```

## Standardize the inputs
Standardizing greatly improves the algorithm

```python
scaled_inputs = preprocessing.scale(unscaled_inputs_equal_priors)
```

## Shuffle the data

```python
shuffled_indices = np.arange(scaled_inputs.shape[0])
np.random.shuffle(shuffled_indices)
shuffled_inputs = scaled_inputs[shuffled_indices]
shuffled_targets = targtes_equal_priors[shuffled_indices]
```

## Split the dataset into train, validation, and test

```python
samples_count = shuffled_inputs.shape[0] # Getting the total number of samples

"""Choose the size of the samples, the old good 80, 10, 10 method"""
train_samples_count = int(0.8*samples_count)
validation_samples_count = int(0.1*samples_count)
test_samples_count = samples_count - train_samples_count - validation_samples_count


train_inputs = shuffled_inputs[:train_samples_count]
train_targets = shuffled_targets[:train_samples_count]

validation_inputs = shuffled_inputs[train_samples_count:train_samples_count+validation_samples_count]
validation_targets = shuffled_targets[train_samples_count:train_samples_count+validation_samples_count]

test_inputs = shuffled_inputs[train_samples_count+validation_samples_count:]
test_targets = shuffled_targets[train_samples_count+validation_samples_count:]

print(np.sum(train_targets), train_samples_count, np.sum(train_targets)/train_samples_count)
print(np.sum(validation_targets), validation_samples_count, np.sum(validation_targets)/validation_samples_count)
print(np.sum(test_targets), test_samples_count, np.sum(test_targets)/test_samples_count)
```
